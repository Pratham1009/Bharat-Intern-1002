```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
sb.set(font_scale=1.2)
```

```python
df = pd.read_csv("../Datasets/STOCKS.csv")
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d%b%y')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Open'] = df['Open'].replace('[$]', '', regex=True)
df['High'] = df['High'].replace('[$]', '', regex=True)
df['Low'] = df['Low'].replace('[$]', '', regex=True)
df['Close'] = df['Close'].replace('[$]', '', regex=True)
df['AdjClose'] = df['AdjClose'].replace('[$]', '', regex=True)
df['Volume'] = df['Volume'].replace('[,]', '', regex=True)
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['Open']= pd.to_numeric(df['Open'])
df['High']= pd.to_numeric(df['High'])
df['Low']= pd.to_numeric(df['Low'])
df['Close']= pd.to_numeric(df['Close'])
df['AdjClose']= pd.to_numeric(df['AdjClose'])
df['Volume']= pd.to_numeric(df['Volume'])
```

```python
df.dtypes
```

```python
df1 = df.set_index('Date')
df1.head()
```

```python
r1 = df[df['Stock']=='IBM']
r2 = df[df['Stock']=='Microsoft']
r3 = df[df['Stock']=='Intel']
```
